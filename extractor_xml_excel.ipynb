```python
# Función para extraer los datos de un archivo XML
def extraer_datos_xml(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    ns = {'cnbv': 'http://www.cnbv.gob.mx'}
    
    # Extracción de los campos requeridos
    numero_oficio = root.findtext('cnbv:Cnbv_NumeroOficio', namespaces=ns)
    solicitud_siara = root.findtext('cnbv:Cnbv_SolicitudSiara', namespaces=ns)
    folio = root.findtext('cnbv:Cnbv_Folio', namespaces=ns)
    area_descripcion = root.findtext('cnbv:Cnbv_AreaDescripcion', namespaces=ns)
    fecha_publicacion = root.findtext('cnbv:Cnbv_FechaPublicacion', namespaces=ns)
    dias_plazo = root.findtext('cnbv:Cnbv_DiasPlazo', namespaces=ns)
    referencia1 = root.findtext('cnbv:Referencia1', namespaces=ns)
    instrucciones = root.findtext('.//cnbv:InstruccionesCuentasPorConocer', namespaces=ns)
    autoridad_nombre = root.findtext('.//cnbv:AutoridadNombre', namespaces=ns)

    persona = root.find('.//cnbv:PersonasSolicitud', namespaces=ns)
    if persona is not None:
        tipo = persona.findtext('cnbv:Persona', namespaces=ns)
        paterno = persona.findtext('cnbv:Paterno', namespaces=ns)
        materno = persona.findtext('cnbv:Materno', namespaces=ns)
        nombre = persona.findtext('cnbv:Nombre', namespaces=ns)
        rfc = persona.findtext('cnbv:Rfc', namespaces=ns)
    else:
        tipo, paterno, materno, nombre, rfc = None, None, None, None, None
    
    return {
        'Cnbv_FechaPublicacion': fecha_publicacion.strip(),
        'Cnbv_NumeroOficio': numero_oficio.strip(),
        'Cnbv_SolicitudSiara': solicitud_siara.strip(),
        'Cnbv_Folio': folio.strip(),
        'Cnbv_AreaDescripcion': area_descripcion.strip(),
        'AutoridadNombre' :autoridad_nombre.strip(),
        'Cnbv_DiasPlazo': dias_plazo.strip(),
        'Referencia1': referencia1.strip(),
        'tipo': tipo.strip() if tipo else None,
        'Paterno': paterno.strip() if paterno else None,
        'Materno': materno.strip() if materno else None,
        'Nombre': nombre.strip() if nombre else None,
        'Rfc': rfc.strip() if rfc else None,
        'InstruccionesCuentasPorConocer': instrucciones.strip() if instrucciones else None,
    }

# Lista de archivos XML
xml_files = [
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (1).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (2).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (3).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (4).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (5).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (6).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (7).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (8).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (9).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (10).XML',
        'C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//31_DE_OCT_SITIAA_PM2//OP_ILI_31_OCT_PM2_SITIAA//XML_31//31_OCT_SITIAA (11).XML',
         ]

# Extracción de datos de todos los archivos XML
data = [extraer_datos_xml(file) for file in xml_files]

# Convertir los datos en un DataFrame de Pandas
df = pd.DataFrame(data)

# Guardar el DataFrame en un archivo Excel
df.to_excel('C://Users//osval//Desktop//Loxical//PM2//SITI_AA_PM2//extracted_data_OP_IL_31_oct.xlsx', index=False)

print("Datos extraídos y guardados en 'extracted_data.xlsx'")
```
